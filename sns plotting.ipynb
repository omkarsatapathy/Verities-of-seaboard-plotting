```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd 
import numpy as np
```

```python
sns_data = sns.load_dataset('penguins')
sns_data
```

```python
sns.scatterplot(x = 'bill_length_mm', y = 'bill_depth_mm', data = sns_data, hue =  'species', palette='Accent', style = 'sex')
plt.show()
```

```python
sns.displot(sns_data['flipper_length_mm'], bins = np.linspace(170,250,9), kde = True, color= 'g')
plt.show()
```

```python
sns.barplot(sns_data['species'] )
plt.show()
```

```python
sns.barplot(x = sns_data['island'], y = sns_data['bill_length_mm'], hue = sns_data['sex'], capsize=0.5)
plt.grid()
plt.show()
```

```python
sns_data2 = sns.load_dataset('anagrams')
sns_data2 = sns_data2.drop(columns=['attnr'], axis=1).head(5)
sns.heatmap(sns_data2, cmap = 'PuOr', annot=True)
```

```python
sns_data3 = sns.load_dataset('tips')
sns_data3.head(3)
```

```python

sns.violinplot(x = sns_data3['day'], y = sns_data3['total_bill'], hue = sns_data3['day'], palette='Accent', linewidth=1)
plt.show()
```

```python
sns.pairplot(sns_data3, vars=['total_bill', 'tip'], hue='sex', kind = 'kde')
plt.show()
```

```python
sns.stripplot(x = sns_data3['day'], y = sns_data3['total_bill'], hue = sns_data3['sex'], 
              palette='pastel', jitter=0.2, marker = '*', size=10)
```
